```python
def plot_all_arrondissements(data):

    fig = plt.figure(figsize=(10,10))

    plt.subplot(3,2,1)
    plt.scatter(data[data['arrondissement'] == 1.0]['surface'], data[data['arrondissement'] == 1.0]['price'])
    plt.title("1er arrondisement")
    plt.xlabel("surface / m2")
    plt.ylabel("price")

    plt.subplot(3,2,2)
    plt.scatter(data[data['arrondissement'] == 2.0]['surface'], data[data['arrondissement'] == 2.0]['price'])
    plt.title("2eme arrondisement")
    plt.xlabel("surface / m2")
    plt.ylabel("price")

    plt.subplot(3,2,3)
    plt.scatter(data[data['arrondissement'] == 3.0]['surface'], data[data['arrondissement'] == 3.0]['price'])
    plt.title("3eme arrondisement")
    plt.xlabel("surface / m2")
    plt.ylabel("price")

    plt.subplot(3,2,4)
    plt.scatter(data[data['arrondissement'] == 4.0]['surface'], data[data['arrondissement'] == 4.0]['price'])
    plt.title("4eme arrondisement")
    plt.xlabel("surface / m2")
    plt.ylabel("price")

    plt.subplot(3,2,5)
    plt.scatter(data[data['arrondissement'] == 10.0]['surface'], data[data['arrondissement'] == 10.0]['price'])
    plt.title("10eme arrondisement")
    plt.xlabel("surface / m2")
    plt.ylabel("price")

    plt.subplot(3,2,6)
    plt.scatter(data['surface'], data['price'])
    plt.title("All of them")
    plt.xlabel("surface / m2")
    plt.ylabel("price")
    
    plt.show()
    
    data.plot.scatter("log-price", "log-surface", c="arrondissement", colormap='viridis')
    
plot_all_arrondissements(data)
```

# Split the train/valid/test  (no validation on this example)

```python
X = np.array([np.ones(data_clean.shape[0]), data_clean['log-surface'], data_clean['arrondissement']]).T
Y = data_clean['log-price']
```

```python
X
```

La constante 1 est ajoutée à chaque observation pour représenter l'ordonnée à l'origine et avoir une écriture vectorielle plus compacte.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

# Compute $\theta$ parameters using the Multivariate OLS formula

```python
thetas = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
```

#### Intercept & Coefficient

```python
thetas
```

#### Predict price for 50m2 in 4th arrondisement

```python
np.expm1( thetas[0] + thetas[1] * np.log1p(50) + thetas[2] * 4 )
```

#### Display regression line along with the data

```python
plt.plot([0,250],[thetas[0], np.expm1(thetas[0]+thetas[1]*np.log1p(250))], linestyle='--', color='r')
plt.scatter(data_clean['surface'], data_clean['price'])
plt.xlabel("log(surface / m2)")
plt.ylabel("log(price)")
```

#### Coefficient of determination

```python
preds = []
for e in X_test:
    pred_log = thetas[0] + thetas[1] * e[1] + thetas[2] * e[2] 
    preds.append(pred_log)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, preds)
```

## Compute $\theta$ parameters using the Sklearn package

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
regression = model.fit(X_train, y_train)
```

#### Coefficient of determination

```python
regression.score(X_test, y_test)
```

#### Intercept & Coefficient

```python
regression.intercept_, regression.coef_[1]
```

#### Predict price for 50m2 in 4th arrondisement

```python
regression.predict(np.array([[1, np.log1p(50), 4]]))
```

```python
print(f"{np.expm1(regression.predict(np.array([[1, np.log1p(50), 4]]))[0]):.2F}€")
```

#### Display regression line along with the data

```python
plt.plot([0,250],[regression.intercept_, np.expm1(regression.intercept_+regression.coef_[1]*np.log1p(250))], linestyle='--', color='r')
plt.scatter(data_clean['surface'], data_clean['price'])
plt.xlabel("surface / m2")
plt.ylabel("price")
```

## Compute ONE model PER `arrondissement`

```python
models = {}

for arr in [1,2,3,4,10]:
    
    # Select & prepare data
    data_arr = data_clean[data_clean['arrondissement']== arr]
    X_arr = np.array([np.ones(data_arr.shape[0]), data_arr['log-surface']]).T
    Y_arr = data_arr['log-price']
    
    # Split data
    X_train_arr, X_test_arr, y_train_arr, y_test_arr = train_test_split(X_arr, Y_arr, test_size=0.2)
    
    # Train model
    model = LinearRegression()
    regression = model.fit(X_train_arr, y_train_arr)
    models[arr] = model
    
    # Evaluate
    score = regression.score(X_test_arr, y_test_arr)
    
    # Display
    print(f"arrondissement:{arr} - score: {score}")
    plt.plot([0,250],[regression.intercept_, np.expm1(regression.intercept_+regression.coef_[1]*np.log1p(250))], linestyle='-', label=arr)
    
plt.scatter(data_clean['surface'], data_clean['price'])
plt.legend()
plt.plot()
```

#### Predict price for 50m2 in 4th arrondisement

```python
models[4].predict(np.array([[1, np.log1p(50)]]))
```

```python
print(f"{np.expm1(models[4].predict(np.array([[1, np.log1p(50)]]))[0]):.2F}€")
```

#### Coefficient of determination when mixing the models

```python
preds = []
for e in X_test:
    pred_log = models[e[2]].predict(np.array([[1, e[1]]]))
    preds.append(pred_log)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, preds)
```

> Ca ne semble pas particulièrement mieux que lorsque l'on prend tous les arrondissements ensemble
